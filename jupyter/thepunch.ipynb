```python
# NOW WRITE DATA INTO A FOLDER WITH TEXT FILES.


url = "https://punchng.com/topics/news/page/{}/"

for page_number in range(1, 4):
    
    print(url.format(page_number))
    response = requests.get(url.format(page_number))
    html_data = soup(response.text) # PARSE HTML DATA INTO BS4

    news_cards = html_data.find_all("div", {"class", "grid-item"}) # FIND ALL DIV TAGS. INCLUDING THE UNWANTED CARDS

    for news_card in news_cards: # NOW MORE CONCISE
        link_tag = news_card.find("a")
        link = link_tag.attrs['href']

        link_response = requests.get(link)
        link_html_data = soup(link_response.text) 
        news = link_html_data.find("div", {"class":"entry-content"})
        
        ###################################
        file_name = F"{news_card.text.strip()}.txt"
        mode = "w"

        file = open(file_name, mode)
        file.write(link_tag.attrs['href'])
        file.write(news.text.strip())
        
        ######################################

        print(link_tag.attrs['href']) # LINK
        print(news_card.text.strip(), end="\n\n") # TITLE
        print(news.text.strip(), end = "\n\n\n") # ACTUAL NEWS ARTICLE
        
    print(f"------------------PAGE {page_number}-----------------")
```

```python
# NOW WRITE DATA INTO A FOLDER WITH TEXT FILES.

file_name = F"PUNCH NEWS.CSV"
mode = "w"
file = open(file_name, mode)


url = "https://punchng.com/topics/news/page/{}/"

for page_number in range(1, 4):
    
    print(url.format(page_number))
    response = requests.get(url.format(page_number))
    html_data = soup(response.text) # PARSE HTML DATA INTO BS4

    news_cards = html_data.find_all("div", {"class", "grid-item"}) # FIND ALL DIV TAGS. INCLUDING THE UNWANTED CARDS

    for news_card in news_cards: # NOW MORE CONCISE
        link_tag = news_card.find("a")
        link = link_tag.attrs['href']

        link_response = requests.get(link)
        link_html_data = soup(link_response.text) 
        news = link_html_data.find("div", {"class":"entry-content"})
        
        news_text = news.text.replace('\n', ' ')
        
        ###################################
        file.write(f"{news_card.text.strip().replace(',', ' ')},{link_tag.attrs['href']},{news_text}\n" )
        ######################################

        print(link_tag.attrs['href']) # LINK
        print(news_card.text.strip(), end="\n\n") # TITLE
        print(news.text.strip(), end = "\n\n\n") # ACTUAL NEWS ARTICLE
        
    print(f"------------------PAGE {page_number}-----------------")
```
