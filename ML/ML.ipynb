```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
x = [2,3,4,5,6,7]
y = [1,2,3,4,5,6] 
```

```python
sns.regplot(x,y) # X AND Y ARE PERFECTLY CORELATED AND HENCE WILL HAVE A PERFECT PLOT
```

```python
x = [6,5,4,3,2,1]
y = [1,2,3,4,5,6] 
```

```python
sns.regplot(x,y) # X AND Y ARE PERFECTLY NEGATIVELY CORELATED AND HENCE WILL HAVE A PERFECT PLOT
```
