```python
x = np.array([1,2,3,4,5,6]).reshape(-1, 1) 

# NOW our single feature data has been reshaped to the required form
y = np.array([1,2,3,4,5,6])**4

y
```

```python
sns.regplot(x,y, order=4)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_modelm
from sklearn.pipeline import make_pipeline

X = np.array([1,2,3,4,5,6]).reshape(-1, 1) 
Y = y.reshape(-1, 1) 

degree=2
polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())

polyreg.fit(X,Y)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(X,Y)
plt.plot(X,polyreg.predict(X),color="black")
plt.title("Polynomial regression with degree "+str(degree))
plt.show()
```

```python
x = range(10)
y = np.array(range(10))**4

X = np.array(x).reshape(-1, 1) 
Y = np.array(y).reshape(-1, 1) 

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))

degree=1

polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X,Y)

ax[0,0].scatter(X,Y)
ax[0,0].plot(X,polyreg.predict(X),color="black")
ax[0,0].set_title("D "+str(degree))
ax[0,0].plot([1,2],[1,2])

degree=2

polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X,Y)

ax[0,1].scatter(X,Y)
ax[0,1].plot(X,polyreg.predict(X),color="black")
ax[0,1].set_title("D "+str(degree))
ax[0,1].plot([1,2],[1,2])

degree=3

polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X,Y)

ax[0,2].scatter(X,Y)
ax[0,2].plot(X,polyreg.predict(X),color="black")
ax[0,2].set_title("D "+str(degree))
ax[0,2].plot([1,2],[1,2])

plt.show()
```

```python
x = [1,2,3,3,5,6,7,7,9,10]
y = np.array(range(10))**4

X = np.array(x).reshape(-1, 1) 
Y = np.array(y).reshape(-1, 1) 

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))
degree=1

for i in range(2):
    for j in range(3):
        

        polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
        polyreg.fit(X,Y)

        ax[i,j].scatter(X,Y)
        ax[i,j].plot(X,polyreg.predict(X),color="black")
        ax[i,j].set_title("D "+str(degree))
        ax[i,j].plot([1,2],[1,2])
        
        degree+=1

plt.show()
```

```python
x = [1,3,3,3,5,6,7,7,9,10]
y = np.array(range(10))**4

X = np.array(x).reshape(-1, 1) 
Y = np.array(y).reshape(-1, 1) 

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))
degree=1

for row in ax:
    for col in row:
        

        polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
        polyreg.fit(X,Y)

        col.scatter(X,Y)
        col.plot(X,polyreg.predict(X),color="black")
        col.set_title("D "+str(degree))
        col.plot([1,2],[1,2])
        
        degree+=1



plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(y, polyreg.predict(X))
```
